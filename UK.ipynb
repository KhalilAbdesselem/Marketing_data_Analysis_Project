```python
import requests
from bs4 import BeautifulSoup

# Step 1: Define the URL of the webpage you want to scrape
url = "https://vidiq.com/youtube-stats/top/country/gb/"
```

```python
# Step 2: Send an HTTP request to the website and get the HTML content
response = requests.get(url)

# Check if the request was successful
if response.status_code == 200:
    html_content = response.text

    # Step 3: Parse the HTML content using BeautifulSoup
    soup = BeautifulSoup(html_content, 'html.parser')

    # Step 4: Find the specific section where the table is located
    section = soup.select_one('body > main > section > div.mx-auto.max-w-\[1172px\].overflow-x-hidden.px-4 > div.flex.flex-col.items-start.justify-between.gap-4.lg\:flex-row.lg\:gap-6 > div.order-1.flex.w-full.flex-col.gap-4.lg\:w-\[calc\(75\%-24px\)\].lg\:gap-6 > div.rounded-\[10px\].bg-vidiq-dark-500.text-white.p-3.md\:p-4.border-vidiq-dark-300.border')
    
    # Step 5: Find the table inside the section (assuming it's a <table> tag)
    table = section.find('table')

    # Step 6: Extract table headers and rows
    if table:
        # Extract table headers
        headers = [header.text.strip() for header in table.find_all('th')]

        # Add "Country" to the headers
        headers.append('Country')

        # Extract table rows (data)
        rows = []
        for row in table.find_all('tr')[1:]:  # Skip the header row
            columns = row.find_all('td')
            row_data = [col.text.strip() for col in columns]

            # Add "United Kingdom" to each row
            row_data.append('United Kingdom')
            rows.append(row_data)

        # Step 7: Print out the extracted data
        print("Table Headers:", headers)
        print("Table Data:")
        for row in rows:
            print(row)

    else:
        print("Table not found in the specified section.")
else:
    print(f"Failed to retrieve the webpage. Status code: {response.status_code}")
```

```python
import pandas as pd
Channels_UK = pd.DataFrame(rows, columns=headers)
```

```python
import requests
from lxml import html
import time
import random

# Define the XPaths for each element
xpaths = {
    'monthly_earning': '/html/body/main/section/div/div[2]/div[2]/div[2]/div[1]/div/div[4]/div/div/p',
    'engagement_rate': '/html/body/main/section/div/div[2]/div[2]/div[2]/div[1]/div/div[5]/div/div/p',
    'video_upload_frequency': '/html/body/main/section/div/div[2]/div[2]/div[2]/div[1]/div/div[6]/div/div/p',
    'average_video_length': '/html/body/main/section/div/div[2]/div[2]/div[2]/div[1]/div/div[7]/div/div/p',
    'category': '/html/body/main/section/div/div[2]/div[2]/div[1]/div[1]/div[1]/div/div[3]/p',
    'total_views_last_month': '/html/body/main/section/div/div[2]/div[2]/div[2]/div[2]/div/div/div[1]/table/tbody/tr[1]/td[3]/span/span'
}

# Function to extract data from a link using XPaths
def extract_data_from_link(link):
    data = {}
    try:
        # Send a GET request to the page
        response = requests.get(link)
        if response.status_code == 200:
            # Parse the HTML content using lxml
            tree = html.fromstring(response.content)

            # Loop through the XPaths and extract the corresponding text content
            for key, xpath in xpaths.items():
                elements = tree.xpath(xpath)  # Extract all elements that match the XPath
                if elements:
                    # Get all the text content from the elements
                    full_text = ' '.join([element.text_content().strip() for element in elements]).strip()
                    data[key] = full_text if full_text else "Text not found"
                else:
                    data[key] = "Element not found"
        else:
            print(f"Failed to retrieve the page. Status code: {response.status_code}")
            for key in xpaths.keys():
                data[key] = "Request failed"
    except Exception as e:
        print(f"Error occurred while processing the link: {link}, Error: {e}")
        for key in xpaths.keys():
            data[key] = "Error occurred"

    return data

# Add columns for the new extracted data to Channels_UK DataFrame
Channels_UK['Monthly Earning'] = ""
Channels_UK['Engagement Rate'] = ""
Channels_UK['Video Upload Frequency'] = ""
Channels_UK['Average Video Length'] = ""
Channels_UK['Category'] = ""
Channels_UK['Total Views Last Month'] = ""

# Loop through each link in the Channels_UK dataframe and extract data
for i, row in Channels_UK.iterrows():
    link = row['Links']
    print(f"Processing link: {link}")
    
    extracted_data = extract_data_from_link(link)
    
    # Update the DataFrame with the extracted data
    Channels_UK.at[i, 'Monthly Earning'] = extracted_data.get('monthly_earning', 'N/A')
    Channels_UK.at[i, 'Engagement Rate'] = extracted_data.get('engagement_rate', 'N/A')
    Channels_UK.at[i, 'Video Upload Frequency'] = extracted_data.get('video_upload_frequency', 'N/A')
    Channels_UK.at[i, 'Average Video Length'] = extracted_data.get('average_video_length', 'N/A')
    Channels_UK.at[i, 'Category'] = extracted_data.get('category', 'N/A')
    Channels_UK.at[i, 'Total Views Last Month'] = extracted_data.get('total_views_last_month', 'N/A')

    # Add a delay between requests to avoid being flagged as a bot
    time.sleep(10 + 10 * random.random())

# Channels_UK DataFrame is now updated with the new columns and extracted data
print("Data extraction completed, Channels_UK is updated.")
```

```python
# Function to extract Channel ID from the link
def extract_channel_id(link):
    # Extract the channel ID from the link
    return link.split('/')[-2]  # Assuming the Channel ID is the second-to-last part of the URL

# Function to extract daily performance data from the link
def extract_daily_performance_data(link):
    try:
        # Send a GET request to the page
        response = requests.get(link)
        if response.status_code == 200:
            # Parse the HTML content using lxml
            tree = html.fromstring(response.content)

            # Find the daily performance table using the provided XPath
            table_xpath = '/html/body/main/section/div/div[2]/div[2]/div[2]/div[2]/div/div/div[1]/table'
            table_elements = tree.xpath(table_xpath)

            # Check if the table exists
            if table_elements:
                # Parse the table into a pandas DataFrame
                table_html = html.tostring(table_elements[0])  # Convert the table element to HTML string
                df_table = pd.read_html(table_html)[0]  # Use pandas to read the HTML table
                return df_table
            else:
                print(f"No table found for link: {link}")
                return pd.DataFrame()  # Return empty DataFrame if no table is found
        else:
            print(f"Failed to retrieve the page. Status code: {response.status_code}")
            return pd.DataFrame()  # Return empty DataFrame if request failed
    except Exception as e:
        print(f"Error occurred while processing the link: {link}, Error: {e}")
        return pd.DataFrame()  # Return empty DataFrame in case of an error

# Initialize an empty DataFrame to store all the performance data
all_performance_data = pd.DataFrame()

# Loop through each link in the Channels_UK dataframe
for i, row in Channels_UK.iterrows():
    link = row['Links']
    print(f"Processing link: {link}")

    # Extract the Channel ID from the link
    channel_id = extract_channel_id(link)
    
    # Extract the daily performance data table
    daily_data_df = extract_daily_performance_data(link)

    if not daily_data_df.empty:
        # Add a Channel_ID column to the extracted table
        daily_data_df['Channel_ID'] = channel_id

        # Append the extracted data to the main DataFrame
        all_performance_data = pd.concat([all_performance_data, daily_data_df], ignore_index=True)

    # Add a delay between requests to avoid being flagged as a bot
    time.sleep(2 + 3 * random.random())


# all_performance_data now contains all the extracted daily performance data with Channel_ID
print("Data extraction completed. Daily performance data stored in DataFrame.")

# Display the first few rows of the DataFrame
print(all_performance_data.head())
```
