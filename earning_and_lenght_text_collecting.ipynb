```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.chrome.options import Options  # Import Options
import time


# Set up Chrome options for headless mode
chrome_options = Options()
chrome_options.add_argument("--headless")  # Run in headless mode
chrome_options.add_argument("--no-sandbox")  # Bypass OS security model (for Linux)
chrome_options.add_argument("--disable-dev-shm-usage")  # Overcome limited resource problems

# Initialize WebDriver with options
service = Service(executable_path='C:\\Users\\khalil\\marketing_data_Analytics_Project\\chromedriver.exe')
driver = webdriver.Chrome(service=service, options=chrome_options)

# Create new columns for earnings and length
df['earning_to_similar'] = None
df['length_to_similar'] = None

# Function to extract earnings and average length
def extract_data(link):
    try:
        driver.get(link)
        wait = WebDriverWait(driver, 10)  # wait up to 10 seconds

        # Extract average video length
        length_xpath = "/html/body/main/section/div/div[2]/div[2]/div[2]/div[1]/div/div[7]/div/div/div[2]/span"
        length_element = wait.until(EC.presence_of_element_located((By.XPATH, length_xpath)))
        length_text = length_element.text.strip()

        # Extract earnings
        earnings_xpath = "/html/body/main/section/div/div[2]/div[2]/div[2]/div[1]/div/div[4]/div/div/div[2]/span"
        earnings_element = wait.until(EC.presence_of_element_located((By.XPATH, earnings_xpath)))
        earnings_text = earnings_element.text.strip()

        return earnings_text, length_text
    
    except Exception as e:
        print(f"Error retrieving data from {link}: {e}")
        return None, None

# Loop through the DataFrame and populate the new columns
for index, row in df.iterrows():
    link = row['Links']
    earnings, length = extract_data(link)
    df.at[index, 'earning_to_similar'] = earnings
    df.at[index, 'length_to_similar'] = length
    time.sleep(2)  # Optional: sleep between requests to avoid throttling

# Close the WebDriver
driver.quit()

# Display the updated DataFrame
print(df)
```

```python
# Connect to the SQL Server database
try:
    with pyodbc.connect(conn_str) as conn:
        cursor = conn.cursor()
        
        # Step 1: Add the new columns to the channels table if they do not exist
        add_columns_query = """
        IF NOT EXISTS (
            SELECT * 
            FROM INFORMATION_SCHEMA.COLUMNS 
            WHERE TABLE_NAME = 'channels' 
            AND COLUMN_NAME = 'earning_to_similar'
        )
        BEGIN
            ALTER TABLE channels 
            ADD earning_to_similar VARCHAR(50);  -- Adjust size as necessary
        END;

        IF NOT EXISTS (
            SELECT * 
            FROM INFORMATION_SCHEMA.COLUMNS 
            WHERE TABLE_NAME = 'channels' 
            AND COLUMN_NAME = 'length_to_similar'
        )
        BEGIN
            ALTER TABLE channels 
            ADD length_to_similar VARCHAR(50);  -- Adjust size as necessary
        END;
        """
        
        cursor.execute(add_columns_query)
        conn.commit()  # Commit the transaction
        print("Columns added successfully.")

        # Step 2: Update the channels table with values from the DataFrame
        for index, row in df.iterrows():
            update_query = """
            UPDATE channels
            SET earning_to_similar = ?, length_to_similar = ?
            WHERE Channel_ID = ?;  -- Assuming you have a unique identifier for the channels
            """
            cursor.execute(update_query, row['earning_to_similar'], row['length_to_similar'], row['Channel_ID'])

        conn.commit()  # Commit the transaction
        print("Table updated successfully.")

except pyodbc.Error as e:
    print("Error while connecting to SQL Server:", e)
```
